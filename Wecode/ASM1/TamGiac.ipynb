```
from math import *
a = float(input())
b = float(input())
c = float(input())

if a+b>c and a-b<c and a+c>b and a-c<b and b+c>a and b-c<a:
  p = (a+b+c)/2
  S = sqrt(p*(p-a)*(p-b)*(p-c))
  if S == int(S):
    S = int(S)
  if a == b and a == c and b == c:
    print('Tam giac deu,','dien tich =',round(S,2))
  elif a == b or a == c or b == c:
    print('Tam gian can,','dien tich =',round(S,2))
  elif a*a == b*b + c*c or b*b == a*a + c*c or c*c == a*a + b*b:
    print('Tam giac vuong,','dien tich =',round(S,2))
  else:
    print('Tam giac thuong,','dien tich =',round(S,2))
else:
  print('Khong phai tam giac')
```
