```
n = int(input())
k = int(input())
p = int(input()) 
q = int(input()) 

vitriA= (p-1) * 2 + (q-1)
trcA = vitriA - k
sauA = vitriA + k
if trcA >= 0:
  pB = trcA//2 + 1
  qB = trcA%2 + 1
  print(pB, qB)
elif sauA <= n-1:
  pB = sauA//2 + 1
  qB = sauA%2 + 1
  print(pB, qB)
else:
  print(-1)
```
