```
def compare(str, com_str):
  for ch in str:
    if ch in com_str:
      return True
  for ch in com_str:
    if ch in str:
      return True
  return False

n = int(input())
for i in range(n):
  str = input()
  com_str = input()
  if str == com_str:
    print('YES')
  else:
    if compare(str, com_str) == False:
      print('NO')
    else:
      print('YES')

    
```
