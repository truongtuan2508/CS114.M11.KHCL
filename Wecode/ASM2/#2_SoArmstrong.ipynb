```
def countDigit(num):
  n = int(num)
  count = 0
  while n > 0:
    n = n // 10
    count = count + 1
  return count

def IsArmstrong(num):
  n = int(num)
  numDigit = int(countDigit(n))
  tmp = n
  sum = 0
  while tmp > 0:
    last = tmp % 10
    tmp = tmp // 10
    sum = sum + pow(last, numDigit)
  if sum == n:
    return True
  return False

n = input()
if IsArmstrong(n) == True:
  print('True')
else:
  print('False')
```
