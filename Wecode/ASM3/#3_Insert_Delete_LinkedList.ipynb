```
 #time_limit_access
import sys

class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
  def __str__(self):
    return str(self.data)
class LinkedList:
  def __init__(self):
    self.head = None
    self.tail = None
  def output(self):
    if self.head is None:
      print("blank")
      return
    else:
      temp_node = self.head
      while temp_node is not None:
        sys.stdout.write(''.join(map(str, str(temp_node.data)))+" ")
        temp_node = temp_node.next
    
  def push_head(self, new_data):
    new_node = Node(new_data)

    if self.head is None:
      self.head = new_node
      self.tail = self.head
    else:
      new_node.next = self.head
      self.head = new_node

  def push_tail(self, new_data):
    new_node = Node(new_data)

    if self.head is None:
      self.head = new_node
      self.tail = self.head
    else:
      self.tail.next = new_node
      self.tail = new_node
  def search(self, val):
    temp_node = self.head
    while temp_node is not None:
      if temp_node.data == val:
        return temp_node
      temp_node = temp_node.next
  def search_be(self,val):
    temp_node = self.head
    while temp_node.next is not None:
      if temp_node.next.data == val:
        return temp_node
      temp_node = temp_node.next

  def push_after(self, val, new_data):
    temp_node = self.search(val)
    if temp_node is not None:
      if temp_node == self.tail:
        self.push_tail(new_data)
      else:
        new_node = Node(new_data)
        new_node.next = temp_node.next
        temp_node.next = new_node
    else:
      self.push_head(new_data)

  def del_val(self,val):    
    if self.head is not None:
      if self.head.data == val:
        self.del_head()
      else:
        temp_node = self.search_be(val)
        if temp_node is not None:
          cur_node = temp_node.next
          if cur_node == self.tail:
            temp_node.next = cur_node.next
            self.tail = temp_node
            del cur_node
          else:
            temp_node.next = cur_node.next
            del cur_node


  
  def check_del(self, val):
    temp_node = self.head
    while temp_node is not None:
      if temp_node.data == val:
        self.del_val(val)
      temp_node = temp_node.next
  
  def del_head(self):
    if self.head is not None:
      if self.head == self.tail:
        del self.head
        self.head = self.tail = None
      else:
        temp_node = self.head
        self.head = self.head.next
        del temp_node

myList = LinkedList()
while True:
  lst = [int(n) for n in sys.stdin.readline().split()]
  if lst[0] == 0:
    myList.push_head(lst[1])
    #print("test", end =" ")
    #myList.output()
  elif lst[0] == 1:
    myList.push_tail(lst[1])
    #print("test", end =" ")
    #myList.output()
  elif lst[0] == 2:
    myList.push_after(lst[1],lst[2])   
    #print("test", end =" ")
    #myList.output()         
  elif lst[0] == 3:
    myList.del_val(lst[1])
    #print("test", end =" ")
    #myList.output()
  elif lst[0] == 4:
    myList.check_del(lst[1])
    #print("test", end =" ")
    #myList.output()
  elif lst[0] == 5:
    myList.del_head()
    #print("test", end =" ")
    #myList.output()
  elif lst[0] == 6:
    break
myList.output()
```
