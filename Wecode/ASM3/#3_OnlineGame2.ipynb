```python
import sys 
listData = set()
while True:
    try:
          lst = [int(n) for n in sys.stdin.readline().split()]
          #lst = [int(n) for n in input().split()]
          if lst[0] == 1:
              listData.add(lst[1])
          elif lst[0] == 2:
            if lst[1] in listData:
              sys.stdout.write("\n"+'1')
            else:
              sys.stdout.write("\n"+'0')
          elif lst[0] == 3:
              listData.discard(lst[1])
          elif lst[0] == 0:
              break
    except:
        break
```
