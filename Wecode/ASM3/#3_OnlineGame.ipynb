```python
import sys 
listData = set()
while True:
  lst = [int(n) for n in sys.stdin.readline().split()]
  #lst = [int(n) for n in input().split()]
  if lst[0] == 1:
      listData.add(lst[1])
  elif lst[0] == 2:
    if lst[1] in listData:
      sys.stdout.write('1'+"\n")
    else:
      sys.stdout.write('0'+"\n")
  elif lst[0] == 0:
    break
```
