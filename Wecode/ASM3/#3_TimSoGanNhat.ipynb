```
n = int(input()) 
myList = list(map(int,input().strip().split()))[:n] 
k, x = map(int,input().split())
from bisect import *
pos = bisect_left(myList, x)
#print("positon" + str(pos))
lst = []

def take_closest(myList,lst, pos_be, pos_af, x, k):
  while k > 0:

    if pos_be < 0 or pos_af >= len(myList):
      break
    before = myList[pos_be]
    after = myList[pos_af]

    #print("check " + str(pos_be) + " " + str(pos_af))
    if after - x < x - before:
      lst.append(after)
      k = k - 1
      pos_af = pos_af + 1
    else:
      lst.append(before)
      k = k - 1
      pos_be = pos_be -1
  if k > 0:
    while k > 0 and pos_be >= 0:
      before = myList[pos_be - 1]
      lst.append(before)
      pos_be = pos_be - 1
      k = k -1
    while k > 0 and pos_af < len(myList):
      after = myList[pos_af]
      lst.append(after)
      pos_af = pos_af + 1
      k = k - 1
  return sorted(lst)
lst = take_closest(myList, lst, pos, pos + 1, x, k)
print(*lst, sep = " ")
```
