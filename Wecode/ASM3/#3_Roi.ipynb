```python
import sys
#h,w = map(int,sys.stdin.readline().split())
h,w = map(int,input().split())
matrix = []
for i in range(h):
  #matrix.append([int(j) for j in sys.stdin.readline().split()])
  matrix.append([int(j) for j in input().split()])
#t,l,b,r = map(int,sys.stdin.readline().split())
t,l,b,r = map(int,input().split())

for i in range(h):
    temp = ['0']*w
    if t <= i and i <= b:
        temp[l:r+1] = matrix[i][l:r+1]
    sys.stdout.write(' '.join(map(str, temp))+"\n")
```
