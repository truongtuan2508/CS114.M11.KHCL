```python
n,m = map(int,input().split())
r,c = map(int,input().split())
x = 0
y = 0
matrix = []
for i in range(n):
   matrix.append([int(j) for j in input().split()])
if n*m == r*c:
  for i in range(r):
    for j in range(c):
      print(matrix[x][y], end = " ")
      y = y + 1
      if y == m:
        y = 0
        x = x + 1
    print()
else:
  for i in matrix:
    for j in i:
        print(j, end=" ")
    print()
```
